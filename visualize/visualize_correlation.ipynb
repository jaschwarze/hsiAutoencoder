```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as patches
import scipy.io as sio
```

```python
mat = sio.loadmat("../test_datasets/Indian_pines_corrected.mat")
img_values = mat["indian_pines_corrected"]
```

```python
# verify implementation on indian pines
img = img_values.reshape(-1, img_values.shape[-1])
corr_mat = np.corrcoef(img, rowvar=False)

plt.matshow(corr_mat)
plt.show()
```

```python
reduced_image_path = "../preprocessing/output/FX10/OK/Kartoffel_Premiere_FX10_9.npy"
new_img_values = np.load(reduced_image_path)
bild = new_img_values.reshape(-1, new_img_values.shape[-1])
new_corr = np.corrcoef(bild, rowvar=False)


fig, (ax1) = plt.subplots()
ax1.matshow(new_corr, interpolation=None, aspect="auto")
ax1.set(xlabel="Bandnummer", ylabel="Bandnummer")
ax1.xaxis.set_label_position("top") 

rectangles = {
    "1-36": patches.Rectangle((0, 0), 35, 35, linewidth=1, edgecolor="r", facecolor="none"),
    "37-71": patches.Rectangle((35, 35), 35, 35, linewidth=1, edgecolor="r", facecolor="none"),
    "72-133": patches.Rectangle((70, 70), 62, 62, linewidth=1, edgecolor="r", facecolor="none"),
    "134-224": patches.Rectangle((132, 132), 91, 91, linewidth=1, edgecolor="r", facecolor="none")
}

for r in rectangles:
    ax1.add_artist(rectangles[r])
    rx, ry = rectangles[r].get_xy()
    cx = rx + rectangles[r].get_width() / 2.0
    cy = ry + rectangles[r].get_height() / 2.0

    ax1.annotate(r, (cx, cy), color="r", weight="bold", fontsize=8, ha="center", va="center")

plt.savefig("segmenteOK", bbox_inches="tight")
```
